# 04 · Seasonal Climatology and the August Anomaly

The paper's headline novel finding is a **late-monsoon (August) salinity peak in the moribund delta** — a signal that challenges the prevailing dry-season-only paradigm.

This notebook:
1. Computes the long-term monthly climatology of NDSI.
2. Removes the seasonal cycle to produce a deseasonalised anomaly series.
3. Isolates the August signal year-by-year.
4. Renders the two key figures: monthly climatology bar chart, August anomaly time-series.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

import warnings
warnings.filterwarnings('ignore')

import pandas as pd

from src.config import DATA_DIR, FIGURES_DIR, RESULTS_DIR
from src.seasonality import (
    monthly_climatology,
    deseasonalise,
    annual_august_anomaly,
)
from src.visualization import plot_monthly_climatology, plot_august_anomaly
```

```python
ts = pd.read_csv(DATA_DIR / 'ndsi_timeseries_2000_2026.csv', parse_dates=['date'])
print(f'Loaded {len(ts)} scenes — {ts["date"].min().date()} to {ts["date"].max().date()}')
```

## 1 · Long-term monthly climatology

```python
clim = monthly_climatology(ts)
print(clim.round(4))

fig_path = FIGURES_DIR / 'fig_monthly_climatology.png'
plot_monthly_climatology(clim, save_to=fig_path)
print(f'\nSaved: {fig_path}')
```

**Interpretation.** The expected dry-season (Jan – Apr) maximum is clearly visible. The secondary August feature — highlighted in the orange bar above — is the novel finding. Three candidate mechanisms documented in `docs/methodology.md`:

1. Mid-monsoon embankment overtopping during peak-discharge events.
2. Reduced upstream freshwater discharge in moribund-delta channels.
3. Capillary rise during brief inter-monsoon dry breaks.

These are hypothesis-generating; causal attribution would require coupled hydrodynamic modelling.

## 2 · Year-by-year August anomalies

```python
aug = annual_august_anomaly(ts)
print(aug.round(5))

aug.to_csv(RESULTS_DIR / 'august_anomaly_by_year.csv', index=False)
```

```python
fig_path = FIGURES_DIR / 'fig_august_anomaly.png'
plot_august_anomaly(aug, save_to=fig_path)
print(f'Saved: {fig_path}')
```
